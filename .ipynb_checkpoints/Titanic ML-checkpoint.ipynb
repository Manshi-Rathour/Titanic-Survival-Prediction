```python
import pandas as pd
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
test_id = test['PassengerId']
```

```python
train.head()
```

```python
test.head()
```

```python
train.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
test.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)
train.head()
```

```python
x = train.drop('Survived', axis='columns')
y = train['Survived']
```

```python
x.isna().sum()
```

```python
test.isna().sum()
```

```python
x.Age = x.Age.fillna(x.Age.mean())
test.Age = test.Age.fillna(test.Age.mean())
test.Fare = test.Fare.fillna(test.Fare.mean())
x.head()
```

```python
x.Sex = x.Sex.map({'male':1, 'female':2})
test.Sex = test.Sex.map({'male':1, 'female':2})
x.head()
```

```python
test.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=30)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
submission_preds = model.predict(test)
```

```python
df = pd.DataFrame({'PassengerId': test_id.values,
                  'Survived': submission_preds})
df.head()
```

```python
df.to_csv('submission.csv', index=False)
```
